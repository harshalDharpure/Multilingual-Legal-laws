```python
import json
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document
```

```python
with open("/DATA/harsh_2311ai14/Akash_QA/MTP02052025/All_rules_child/english_laws.txt", "r", encoding="utf-8", errors="replace") as f:
    legal_text = f.read()

print(legal_text)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = text_splitter.create_documents([legal_text])
print("HII")
embedding_model = HuggingFaceEmbeddings(model_name="intfloat/e5-base-v2")
vectorstore = FAISS.from_documents(documents, embedding_model)
```

```python
# ----- Step 1: Load and Chunk Legal Text (.txt) -----




# ----- Step 3: Process and Save One-by-One -----

input_path = "/DATA/harsh_2311ai14/Akash_QA/MTP_final/data_final/test_data.jsonl"
output_path = "/DATA/harsh_2311ai14/Akash_QA/MTP02052025/All_rules_child/rag_retrieved_questions.jsonl"

with open(input_path, "r", encoding="utf-8") as fin, open(output_path, "a", encoding="utf-8") as fout:
    for line in fin:
        try:
            q = json.loads(line)
            query = f"{q['Question']} Options: A: {q['A']}, B: {q['B']}, C: {q['C']}, D: {q['D']}"
            print(query)
            docs = vectorstore.similarity_search(query, k=5)
            context = "\n".join([doc.page_content for doc in docs])
            print("Hii")
            output = {
                "Passage": q["Passage"],
                "Question": q["Question"],
                "A": q["A"],
                "B": q["B"],
                "C": q["C"],
                "D": q["D"],
                "Correct Answer": q["Correct Answer"],
                "Reasoning": q["Reasoning"],
                "retrieved_context": context
            }

            fout.write(json.dumps(output, ensure_ascii=False) + "\n")
        
        except Exception as e:
            print(f"Error processing line: {line}\nError: {e}")
```
